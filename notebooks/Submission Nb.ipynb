```python
import os
import datetime
import pandas as pd
import numpy as np

from keras.initializers import he_uniform
from keras.models import Sequential, load_model
from keras.optimizers import Adam, RMSprop
from keras.regularizers import l2
from keras.layers import Dense, Dropout, BatchNormalization, Activation
from keras.utils import np_utils
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import LearningRateScheduler, EarlyStopping, ModelCheckpoint

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import confusion_matrix

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

seed=11
np.random.seed(seed)
```

```python
payload_dir = '../payload/'
eval_file = 'normal-eval-payload.npz'

eval_path = os.path.join(payload_dir, eval_file)
eval_archive = np.load(eval_path)
eval_set = eval_archive['eval_set'].reshape(-1,28,28,1)
```

## Model

```python
models_dir = '../models/'
model_name = 'ConvNet.0.02.hdf5'

modelpath = os.path.join('../models/ConvNet.01-0.02.hdf5')
```

```python
model = load_model(modelpath)
```

```python
eval_set.shape
```

```python
pred = model.predict_classes(eval_set)
```

```python
imid = np.arange(1,len(pred)+1)


subm = pd.DataFrame(np.c_[imid, pred], columns=['ImageId', 'Label']).to_csv(index=False)

with open('../submissions/'+model_name+'.csv', 'w') as f:
    f.write(subm)
```
